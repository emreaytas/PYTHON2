```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numpydizisi1 = np.linspace(0,10,20)
```

```python
numpydizi2 = numpydizisi1 ** 2
```

```python
(fig,axs) = plt.subplots()
axs.plot(numpydizisi1,numpydizi2,"g*-",label = "kare")
axs.plot(numpydizi2,numpydizisi1,"b",label = "degisik")
axs.set_xlabel("x ekseni")
plt.show()
# renk kodlarını rgb hex code ile bulabiliriz. mesela color = "#3A95A8" yapabiliriz.
# eğer transparan olsun istersek aplha = sayi gibi bir şey yaparız 1 olunca mat 0.00... olunca ise saydam olur çizgimiz.
```

```python
(fig,axs) = plt.subplots()
axs.plot(numpydizisi1,numpydizisi1 + 22,color = "#3A95A8",alpha = 0.5,linewidth = 2.0,linestyle = "--") # linestyle = "--" ile biz çizgi şeklini belirleriz.   
axs.plot(numpydizisi1,numpydizisi1 + 32,color = "g",alpha = 0.2,linewidth = 6.0,linestyle = ":")  # linewidth = sayi ile kalınlık belirleyebiliriz.
plt.show()
```

```python
(fig,axs) = plt.subplots()
axs.plot(numpydizisi1,numpydizisi1 + 22,color = "#3A95A8",alpha = 0.5,linewidth = 2.0,linestyle = "--") # linestyle = "--" ile biz çizgi şeklini belirleriz.   
axs.plot(numpydizisi1,numpydizisi1 + 32,color = "g",alpha = 0.2,linewidth = 6.0,linestyle = "-.",marker =  "o",markersize = 8,markerfacecolor = "red")  # linewidth = sayi ile kalınlık belirleyebiliriz.
plt.show()
# marker = "sembol" ile çizgi üzerine ekleme yapabiliriz. markerler ile çizgi üzerine bir şey ekledik.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline # plt.show() ile grafikleri g√∂sterebiliriz.
```

```python
yaslistesi = [10,20,30,30,30,40,50,60,70,75]
kilolistesi = [20,60,80,85,86,87,70,90,95,90]
npyas = np.array(yaslistesi)
npkilo = np.array(kilolistesi)
```

```python

plt.plot(npyas,npkilo,color = "r",label = "cizgi")  # √∂nce x ekseni elemanlarƒ± sonra ise y ekseni elemanlarƒ± sonra renklendirme label vs alƒ±r.
plt.xlabel("x ekseni") # x ekseni adlandƒ±rma
plt.ylabel("y ekseni")
plt.title("GRAFƒ∞K") # tabloya isim verme 
plt.legend(loc = 2) # bilgilendirme tablosunun eksende hangi b√∂lgede olduƒüu.
plt.show() # g√∂ster emri
```

```python
newnumpy = np.linspace(0,10,20)
plt.plot(newnumpy,newnumpy ** 3,color = "g",label="cubic")
plt.xlabel("x ekseni") # x ekseni adlandƒ±rma
plt.ylabel("y ekseni")
plt.title("GRAFƒ∞K") # tabloya isim verme 
plt.legend(loc = 2) # bilgilendirme tablosunun eksende hangi b√∂lgede olduƒüu.
plt.show() # g√∂ster emri
```

```python
newnumpy = np.linspace(0,10,20)
plt.plot(newnumpy,newnumpy ** 3,"g--",label="cubic") # "g--" √ßizgi bir gragik olur
plt.xlabel("x ekseni") # x ekseni adlandƒ±rma
plt.ylabel("y ekseni")
plt.title("GRAFƒ∞K") # tabloya isim verme 
plt.legend(loc = 2) # bilgilendirme tablosunun eksende hangi b√∂lgede olduƒüu.
plt.show() # g√∂ster emri
```

```python
newnumpy = np.linspace(0,10,20)
plt.plot(newnumpy,newnumpy ** 3,"g*-",label="cubic") # "g*-"  kesik kesik yƒ±ldƒ±zlardan olu≈üur.  her yƒ±ldƒ±z bir ke≈üisim noktasƒ± olur.
plt.xlabel("x ekseni") # x ekseni adlandƒ±rma
plt.ylabel("y ekseni")
plt.title("GRAFƒ∞K") # tabloya isim verme 
plt.legend(loc = 2) # bilgilendirme tablosunun eksende hangi b√∂lgede olduƒüu.
plt.show() # g√∂ster emri
```

```python
plt.subplot(1,2,1) # subplot alt grafik demek. √∂nce ka√ß sƒ±ra olacak. sonra ka√ß tane grafik sonra hangisini √ßizeceƒüiz belirtiriz.
plt.plot(newnumpy,newnumpy ** 2,"r*-",label = "quadrotic")
plt.subplot(1,2,1).set_xlabel("X")
plt.subplot(1,2,1).set_ylabel("Y")
plt.subplot(1,2,1).set_title("Grafik 1")
plt.legend()

plt.subplot(1,2,2) # en sondaki 2 ile biz ikinci olanƒ± √ßizeceƒüimizi belirttik.
plt.plot(newnumpy,newnumpy ** 3,"r*-",label = "cubic")
plt.subplot(1,2,2).set_xlabel("X")
plt.subplot(1,2,2).set_ylabel("Y")
plt.subplot(1,2,2).set_title("Grafik 2")
plt.legend()
plt.savefig("emre1.png")
plt.show()
# birden fazla plot ile i≈ülem yapabiliriz. tek bir sayfa √ºzerinde. bunun i√ßin subplot vs kullanmak lazƒ±m.
# shift + tab yaparsak bir metotun vs infosunu g√∂rebiliriz d√∂k√ºmana ula≈üabiliriz.
# subplot ile biz belirleriz hangi grafik sonra plt √ºzerinden belli olana i≈ülem yapabiliriz.
```

```python
benimfigure = plt.figure() # figur olu≈üturma. nedir = bo≈ü bir figur olu≈üturur. ama biz geni≈ülik vs belirleyebiliriz.
figureaxes = benimfigure.add_axes([0.2,0.2,0.9,0.9]) # arg√ºmanlarƒ± bir liste i√ßerisinde veririz.
figureaxes.plot(newnumpy,newnumpy ** 2,"g")
figureaxes.set_xlabel("X ekseni.")
figureaxes.set_ylabel("Y ekseni.")
figureaxes.set_title("figur")
plt.show()

```

```python
benimfigure = plt.figure() # figur olu≈üturma. nedir = bo≈ü bir figur olu≈üturur. ama biz geni≈ülik vs belirleyebiliriz.
figureaxes = benimfigure.add_axes([0.4,0.4,0.3,0.3]) # arg√ºmanlarƒ± bir liste i√ßerisinde veririz.
figureaxes.plot(newnumpy,newnumpy ** 2,"g")
figureaxes.set_xlabel("X ekseni.")
figureaxes.set_ylabel("Y ekseni.")
figureaxes.set_title("figur")
plt.show()
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.3,0.3,0.3,0.3])
eksen1.plot(newnumpy,newnumpy ** 2,"g")
eksen1.set_xlabel("X")
eksen1.set_ylabel("Y")
eksen1.set_title("Tƒ∞TLE 1")
eksen2.plot(newnumpy ** 2,newnumpy,"b")
eksen2.set_xlabel("X")
eksen2.set_ylabel("Y")
eksen2.set_title("Tƒ∞TLE2")
plt.show()
```

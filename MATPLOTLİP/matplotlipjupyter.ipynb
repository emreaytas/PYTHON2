```python
import numpy as np
import matplotlib.pyplot as plt   # matplotlib içerisinden pyplot'u çağırırız. bunu kullanacağız.  as ile isimlendirme yaparız..
```

```python
%matplotlib inline 
# plt.show() ile grafikleri gösterebiliriz.
# eğer istersek matplotlib sitesinden kendimize göre veri görselleştirme araçlarına bakabiliriz.
```

```python
yaslistesi = [10,20,30,30,30,40,50,60,70,75]
kilolistesi = [20,60,80,85,86,87,70,90,95,90]
npyas = np.array(yaslistesi)
npkilo = np.array(kilolistesi)
```

```python

plt.plot(npyas,npkilo,color = "r",label = "cizgi")  # önce x ekseni elemanları sonra ise y ekseni elemanları sonra renklendirme label vs alır.
plt.xlabel("x ekseni") # x ekseni adlandırma
plt.ylabel("y ekseni")
plt.title("GRAFİK") # tabloya isim verme 
plt.legend(loc = 2) # bilgilendirme tablosunun eksende hangi bölgede olduğu.
plt.show() # göster emri
```

```python
newnumpy = np.linspace(0,10,20)
plt.plot(newnumpy,newnumpy ** 3,color = "g",label="cubic")
plt.xlabel("x ekseni") # x ekseni adlandırma
plt.ylabel("y ekseni")
plt.title("GRAFİK") # tabloya isim verme 
plt.legend(loc = 2) # bilgilendirme tablosunun eksende hangi bölgede olduğu.
plt.show() # göster emri
```

```python
newnumpy = np.linspace(0,10,20)
plt.plot(newnumpy,newnumpy ** 3,"g--",label="cubic") # "g--" çizgi bir gragik olur
plt.xlabel("x ekseni") # x ekseni adlandırma
plt.ylabel("y ekseni")
plt.title("GRAFİK") # tabloya isim verme 
plt.legend(loc = 2) # bilgilendirme tablosunun eksende hangi bölgede olduğu.
plt.show() # göster emri
```

```python
newnumpy = np.linspace(0,10,20)
plt.plot(newnumpy,newnumpy ** 3,"g*-",label="cubic") # "g*-"  kesik kesik yıldızlardan oluşur.  her yıldız bir keşisim noktası olur.
plt.xlabel("x ekseni") # x ekseni adlandırma
plt.ylabel("y ekseni")
plt.title("GRAFİK") # tabloya isim verme 
plt.legend(loc = 2) # bilgilendirme tablosunun eksende hangi bölgede olduğu.
plt.show() # göster emri
```

```python
plt.subplot(1,2,1) # subplot alt grafik demek. önce kaç sıra olacak. sonra kaç tane grafik sonra hangisini çizeceğiz belirtiriz.
plt.plot(newnumpy,newnumpy ** 2,"r*-",label = "quadrotic")
plt.subplot(1,2,1).set_xlabel("X")
plt.subplot(1,2,1).set_ylabel("Y")
plt.subplot(1,2,1).set_title("Grafik 1")
plt.legend()

plt.subplot(1,2,2) # en sondaki 2 ile biz ikinci olanı çizeceğimizi belirttik.
plt.plot(newnumpy,newnumpy ** 3,"r*-",label = "cubic")
plt.subplot(1,2,2).set_xlabel("X")
plt.subplot(1,2,2).set_ylabel("Y")
plt.subplot(1,2,2).set_title("Grafik 2")
plt.legend()
plt.savefig("emre1.png")
plt.show()
# birden fazla plot ile işlem yapabiliriz. tek bir sayfa üzerinde. bunun için subplot vs kullanmak lazım.
# shift + tab yaparsak bir metotun vs infosunu görebiliriz dökümana ulaşabiliriz.
# subplot ile biz belirleriz hangi grafik sonra plt üzerinden belli olana işlem yapabiliriz. plt.subplot(1,2,2) mesela bundan sonra ikinci grafik üzerinde işlem yapmaya başlarız.s
```

```python
benimfigure = plt.figure() # figur oluşturma. nedir = boş bir figur oluşturur. ama biz genişlik vs belirleyebiliriz.
figureaxes = benimfigure.add_axes([0.2,0.2,0.9,0.9]) # argümanları bir liste içerisinde veririz.
figureaxes.plot(newnumpy,newnumpy ** 2,"g")
figureaxes.set_xlabel("X ekseni.")
figureaxes.set_ylabel("Y ekseni.")
figureaxes.set_title("figur")
plt.show()

```

```python
(fig,axs) = plt.subplots(nrows = 1,ncols = 2) # bir tuple döndürür. figur ve axis bilgisi çıkarır. birden fazla eksen olursa bir ndarray döner

```

```python
(fig,axs) = plt.subplots(nrows = 1,ncols = 2) # bir tuple döndürür. figur ve axis bilgisi çıkarır. birden fazla eksen olursa bir ndarray döner eksenler döndürür.
for eksen in axs:
    eksen.plot(newnumpy,newnumpy ** 2, label = "kare",color = "g")
    eksen.set_xlabel("X")
    eksen.set_ylabel("Y")
    eksen.set_title("TİTLE")
plt.tight_layout() # birbirine girmesini engeller yazıların falan    
```

```python
(fig,axs) = plt.subplots(nrows = 1,ncols = 2) # bir tuple döndürür. figur ve axis bilgisi çıkarır. birden fazla eksen olursa bir ndarray döner eksenler döndürür.
axs[0].plot(newnumpy,newnumpy ** 2, label = "kare",color = "g")
axs[0].set_xlabel("XX")
axs[0].set_ylabel("YY")
axs[0].set_title("title1")
axs[1].plot(newnumpy,newnumpy ** 3, label = "cubic",color = "b")
axs[1].set_xlabel("xx")
axs[1].set_ylabel("yy")
axs[1].set_title("title2")
plt.tight_layout()
plt.show()
```

```python
yenifigur = plt.figure() # figur üzerinde birden fazla plot oluşturabiliriz. figsize = (a,b) vermezsek eğer o zaman otomatik olarak sayfaya göre ayarlar.
yenieksen = yenifigur.add_axes([0.2,0.1,0.9,0.9])
yenieksen.plot(newnumpy,newnumpy ** 2,label = "kare") # label ile çizgiye isim veririz
yenieksen.plot(newnumpy,newnumpy ** 4,label = "dört")
yenieksen.plot(newnumpy,newnumpy ** 5,label = "bes")
yenieksen.legend(loc = 2)
plt.show()
```

```python
yenifigur = plt.figure(figsize = (4,4)) # figur üzerinde birden fazla plot oluşturabiliriz. figsize() ile weight ve height belirleyebilriz bir tuple içerisinde veririz.
yenieksen = yenifigur.add_axes([0.2,0.1,0.9,0.9])
yenieksen.plot(newnumpy,newnumpy ** 2,label = "kare") # label ile çizgiye isim veririz
yenieksen.plot(newnumpy,newnumpy ** 4,label = "dört")
yenieksen.plot(newnumpy,newnumpy ** 5,label = "bes")
yenieksen.legend(loc = 2)
plt.show()
```

```python
yenifigur = plt.figure(figsize = (4,4)) # dpi = 100 gibi dpi ile kalite belirtebiliriz.
yenieksen = yenifigur.add_axes([0.2,0.2,0.7,0.8])
yenieksen.plot(newnumpy,newnumpy ** 2,label = "kare") # label ile çizgiye isim veririz
yenieksen.plot(newnumpy,newnumpy ** 3,label = "küp")
yenieksen.plot(newnumpy,newnumpy ** 4,label = "dört")
yenieksen.plot(newnumpy,newnumpy ** 5,label = "bes")
yenieksen.set_xlabel("X")
yenieksen.set_ylabel("Y")
yenieksen.set_title("baslik")
yenieksen.legend(loc = 2)

plt.savefig("emre2.png",dpi = 100) # kaydederken dpi belirleyebiliriz. illa çalışırken atamamıza gerek yok.
plt.show()
```
